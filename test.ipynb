```python
import pandas as pd

df = pd.read_csv("archive/imdb-movies-dataset.csv")
# Giả sử df là DataFrame của bạn
# Chuyển Rating sang float và loại bỏ giá trị không hợp lệ
df['Rating'] = pd.to_numeric(df['Rating'], errors='coerce')

# Lấy giá trị rating cao nhất
max_rating = df['Rating'].max()

# Lấy tất cả các phim có rating bằng max_rating
top_movies = df[df['Rating'] == max_rating]

# Hiển thị
top_movies[['Title', 'Rating', 'Year', 'Genre']]
```

```python
import pandas as pd

# --- Load dữ liệu ---
df_main = pd.read_csv("archive\imdb-movies-dataset.csv")  # file chính
df_fix  = pd.read_csv("archive\imdb_top_1000.csv")   # file sửa

# --- Merge chỉ để lấy cột Released_Year ---
df_merged = df_main.merge(
    df_fix[['Series_Title', 'Released_Year']],  # chỉ lấy 2 cột cần thiết
    left_on='Title',
    right_on='Series_Title',
    how='left'  # giữ tất cả phim của df_main, bỏ qua phim fix không có trong main
)

# --- Thay Year bằng Released_Year nếu có ---
df_merged['Year'] = df_merged['Released_Year'].combine_first(df_merged['Year'])

# --- Xóa cột Series_Title và Released_Year dư thừa ---
df_merged.drop(columns=['Series_Title', 'Released_Year'], inplace=True)

# --- Kiểm tra kết quả ---
print(df_merged.head())

df_merged.to_csv("archive\imdb-movies-dataset.csv", index=False)
```

```python
import pandas as pd

# Giả sử file gốc là 'movies.csv'
file_path = 'archive\imdb-movies-dataset.csv'

# Load dữ liệu
df = pd.read_csv(file_path)

# Chuyển Rating sang số
df['Rating'] = pd.to_numeric(df['Rating'], errors='coerce')

# Sửa Rating của phim 'August' xuống 5
df.loc[df['Title'] == 'August', 'Rating'] = 5

# Lưu lại trực tiếp vào file gốc
df.to_csv(file_path, index=False)

print("Rating của phim 'August' đã được cập nhật xuống 5.")
```

```python
import pandas as pd

file_path = 'archive\imdb-movies-dataset.csv'   # đổi nếu tên file khác

# Load dữ liệu
df = pd.read_csv(file_path)

# Chuyển Rating sang số (phòng trường hợp có lỗi)
df['Rating'] = pd.to_numeric(df['Rating'], errors='coerce')

# Cập nhật rating theo yêu cầu
df.loc[df['Title'] == 'Just Now Jeffrey', 'Rating'] = 9.0
df.loc[df['Title'] == 'Los Frikis', 'Rating'] = 6.6

# Lưu vào chính file gốc
df.to_csv(file_path, index=False)

print("Đã cập nhật rating cho 2 phim.")
```
