```python
import pandas as pd

# --- Load dữ liệu ---
df_main = pd.read_csv("archive\imdb-movies-dataset.csv")  # file chính
df_fix  = pd.read_csv("archive\imdb_top_1000.csv")   # file sửa

# --- Merge chỉ để lấy cột Released_Year ---
df_merged = df_main.merge(
    df_fix[['Series_Title', 'Released_Year']],  # chỉ lấy 2 cột cần thiết
    left_on='Title',
    right_on='Series_Title',
    how='left'  # giữ tất cả phim của df_main, bỏ qua phim fix không có trong main
)

# --- Thay Year bằng Released_Year nếu có ---
df_merged['Year'] = df_merged['Released_Year'].combine_first(df_merged['Year'])

# --- Xóa cột Series_Title và Released_Year dư thừa ---
df_merged.drop(columns=['Series_Title', 'Released_Year'], inplace=True)

# --- Kiểm tra kết quả ---
print(df_merged.head())

df_merged.to_csv("archive\imdb-movies-dataset.csv", index=False)
```

```python
import pandas as pd
import re
import numpy as np


def increase_poster_size_revised(url):
    """
    Sửa đổi URL bằng cách loại bỏ các tham số giới hạn kích thước 
    (Ví dụ: _UY209_CR0,0,140,209_AL_)
    """
    if pd.isna(url) or not isinstance(url, str):
        return url
        
    # Pattern: Tìm kiếm chuỗi bắt đầu từ _V1 cho đến trước _AL
    # Đây là pattern phức tạp vì cấu trúc Amazon/IMDb không hoàn toàn nhất quán.
    # Phương pháp đơn giản nhất là xóa tất cả các tham số nằm giữa ._V1 và phần mở rộng file (ví dụ: .jpg)
    
    # 1. Tách chuỗi tại ._V1
    parts = url.split('._V1')
    if len(parts) < 2:
        return url # Trả về nếu không tìm thấy mẫu
        
    # 2. Lấy phần cuối (chứa tham số và đuôi file)
    end_part = parts[1]
    
    # 3. Tìm đuôi file (ví dụ: .jpg) trong phần cuối
    match = re.search(r'(\.(jpg|png|jpeg))', end_part, re.IGNORECASE)
    
    if match:
        # Giữ lại đuôi file (ví dụ: .jpg)
        file_extension = match.group(0)
        
        # Gộp lại: Phần đầu + ._V1 + .jpg (hoặc .png)
        cleaned_url = parts[0] + '._V1' + file_extension
        return cleaned_url
        
    return url 


# Lưu URL gốc để kiểm tra so sánh
df['Poster_Original'] = df['Poster']

# Áp dụng hàm sửa URL đã được tối ưu
df['Poster'] = df['Poster'].apply(increase_poster_size_revised)

print("\n--- KẾT QUẢ SỬA URL MỚI (5 Mẫu Đầu Tiên) ---")
print(df[['Poster_Original', 'Poster']].head())

# Kiểm tra mẫu bạn cung cấp
example_url = "https://m.media-amazon.com/images/M/MV5BYWRkZjJiODEtM2IwZi00ZjM1LWEyOTUtOThjMDk3YThjZDUzXkEyXkFqcGdeQXVyMTUzMTg2ODkz._V1_UY209_CR0,0,140,209_AL_.jpg"
print("\nVí dụ gốc của bạn:", example_url)
print("Ví dụ mới sau khi sửa:", increase_poster_size_revised(example_url))

```

```python
import pandas as pd
import os

# 1. Định nghĩa tên file
file_path = 'recommend_system.csv'

# 2. Kiểm tra file có tồn tại không trước khi đọc
if not os.path.exists(file_path):
    print(f"Lỗi: Không tìm thấy file tại đường dẫn: {file_path}")
else:
    try:
        # 3. Đọc file vào DataFrame. Giả định file là CSV, nếu là Excel dùng pd.read_excel
        df = pd.read_csv(file_path)
        
        # 4. Kiểm tra cột 'Poster_Original' có tồn tại không
        if 'Poster_Original' in df.columns:
            # 5. Xóa cột 'Poster_Original'
            # axis=1 chỉ định rằng chúng ta đang xóa một cột
            # inplace=True sẽ thực hiện thay đổi trực tiếp trên DataFrame df
            df.drop('Poster_Original', axis=1, inplace=True)
            
            print("Đã xóa cột 'Poster_Original' thành công.")
            
            # 6. Lưu DataFrame đã chỉnh sửa ĐÈ lên file gốc
            # index=False ngăn việc ghi chỉ mục (index) của DataFrame vào file
            df.to_csv(file_path, index=False)
            
            print(f"Đã lưu đè dữ liệu mới vào file: {file_path}")
            
        else:
            print("Cột 'Poster_Original' không tồn tại trong file. Không có thay đổi nào được thực hiện.")
            
    except Exception as e:
        print(f"Đã xảy ra lỗi trong quá trình xử lý file: {e}")
```
